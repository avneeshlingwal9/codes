```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def strToList(el):
    curr = ""
    lst = []
    for i in el:
        if(i == " "):
            lst.append(int(curr))
            curr = ""
        else:
            curr+=i

    return lst
```

```python
def fileOpen(name):
    file = open(name,"r")
    return file.readline()

```

```python
def plot(size, file):
    curr = fileOpen(file)
    curr = strToList(curr)
    plt.plot(size,curr, label ='2x', color = 'red', linewidth = 2 , linestyle = '--' )
    plt.xlabel("Size of the array")
    plt.ylabel("Number of operations")
    plt.show()
```

```python
size = fileOpen("numberOfElements.csv")
size = strToList(size)
```

```python
plot(size,"bestcase.csv")

```

```python
plot(size, "worstcase.txt")
```

```python
plot(size, "averagecase.txt")
```
