```python
# Libraries helpful in plotting graphs.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''
Function:  str_to_list(str)

Description : 

    Given a string str   converts it to a list. 

Input Parameters:

    string str: The string containing number of comparisons separated by a space. 

Output Parameters: 

    list: A string converted into a list of numbers.



'''
def str_to_list(el):
    curr = ""
    lst = []
    for i in el:
        if(i == " "):
            lst.append(int(curr))
            curr = ""
        else:
            curr+=i

    return lst
```

```python
'''
Function:  file_open(filename)

Description : 

    Opens a file filename and returns its contents.

Input Parameters:

    string filename: Name of the file to be opened. 

Output Parameters: 

    A string containing file data.



'''


def file_open(name):
    file = open(name,"r")
    return file.readline()
```

```python
'''
Function:  plot(size , filename)

Description : 

    Given a list size and a filename. It opens the file and plots it, in comparison with list size. 

Input Parameters:

    list size: A list containing sizes of all different arrays.

    string filename: A file containing comparisons for those respective different sizes of arrays.

Output Parameters: 

    A graph



'''


def plot(size, filename):
    curr = file_open(filename)
    curr = str_to_list(curr)
    plt.plot(size,curr, label ='2x', color = 'red', linewidth = 2 , linestyle = '--' )
    plt.xlabel("Size of the array")
    plt.ylabel("Number of operations")
    plt.show()
```

```python
# Storing the size of different arrays in a file. 
size = file_open("size.txt")
size = str_to_list(size)
```

Best case comparisons

```python
#Plotting for best case comparisons. 

plot(size, "bestCaseInsertionSort.txt")
```

Worst case comparisons:

```python
# Plotting for worst case comparisons

plot(size, "worstCaseInsertionSort.txt")
```

Average Case comparisons:

```python
#  Plotting for average case. 

plot(size, "averageCaseInsertionSort.txt")
```
